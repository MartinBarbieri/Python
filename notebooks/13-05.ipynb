# **Entidades bancarias en CABA y su distribución por comunas**

```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import numpy as np


df1 = pd.read_csv("../data/accesibilidad-en-bancos.csv",sep=";",engine="python" )
x_values = df1['NOMBRE'].unique()
y_values = df1['NOMBRE'].value_counts().tolist()
y_values.reverse()
x_values= np.flip(x_values)
# gráfico 1
plt1.bar(x_values,y_values)
plt1.title('Cantidad de sucursales de cada banco')
plt1.xlabel('Nombre')
plt1.ylabel('Cantidad')
plt1.xticks(rotation=63,fontsize=11)
plt1.show()

x_values = df1['COMUNA'].value_counts().index.tolist() + ["COMUNA 4"]
y_values =  df1['COMUNA'].value_counts().tolist() + ["0"]
y_values.reverse()
x_values= np.flip(x_values)
# gráfico 2
plt2.plot(x_values,y_values)
plt2.title('Cantidad de bancos por comuna')
plt2.xlabel('Comunas')
plt2.ylabel('Cantidad')
plt2.xticks(rotation=63,fontsize=11)
plt2.show()
```

### **Conclusiones**

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Viendo ambos gráficos puedo concluir que Santander Rio es la cadena
bancaria con mayor presencia en la Ciudad Autónoma de Buenos Aires seguido
de cerca por el Banco Patagonia. Se puede destacar que de todas las sucursales del primer mencionado
solo 6 cuentan con rampa para personas con discapacidad.
Sin embargo, la diferencia entre cuántas sucursales de cada banco hay
no es tanta como la presencia de estos en las comunas.
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;La comuna 8 tiene ampliamente mayor cantidad de sucursales. Esto podría llegar a
ser ya que los barrios que la integran (Villa Soldati, Villa Riachuelo y
Villa Lugano) están en el límite entre la provincia y la capital. Opuesto a esto, en la comuna
4 no se cuenta con ninguna sucursal.
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Una variable que el CSV no posee (pero se podría tener en cuenta) es si estas
filiales cuentan con caja de ahorro o no. De esta forma se llegaría a inferir
el estado económico de la poblacion. Si la mayoría posee caja fuerte
apunta mas a una población con un buen pasar económico que tiene ingresos
sobrantes para ahorrar. En cambio, si no sucede esto, quiere decir que el país estaría
atravesando una crisis y, en consecuencia, la plata con la que cuentan los
ciudadanos es la justa.
